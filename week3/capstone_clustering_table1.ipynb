### Scraping of the html data source

```python

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df_list = pd.read_html(url)
df = df_list[0]
df
```

### Filtering the rows with "Not assigned" Borough cells 

```python
df = df.drop(df[df['Borough'] == 'Not assigned'].index)
df
```

### Verifying if there's some "Neighbourhood" with the content "Not Assigned"

```python
df[df["Neighbourhood"] == "Not assigned"]
```

### Joining the neighbourhoods with the same postal code

```python
df = df[["Postal Code", "Borough", "Neighbourhood"]].loc[df["Postal Code"].index]
df
```
