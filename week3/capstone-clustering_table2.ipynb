### Previous data work

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df_list = pd.read_html(url)
df = df_list[0]
df = df.drop(df[df['Borough'] == 'Not assigned'].index)
df[df["Neighbourhood"] == "Not assigned"]
df = df[["Postal Code", "Borough", "Neighbourhood"]].loc[df["Postal Code"].index]
df = df.sort_values("Postal Code")
df.reset_index(inplace=True, drop=True) 
df
```

### Readind the csv coordnates file in a new dataframe

```python
df2 = pd.read_csv("Geospatial_Coordinates.csv")
df2
```

### Renaming the common Postal Code column

```python
df2.columns = ["PCode", "Latitude", "Longitude"]
df2.head()
```

### Joining the dataframes and filtering repeated column

```python
dframes = [df, df2]
df3 = pd.concat(
    dframes,
    axis=1,
    join="outer",
    ignore_index=False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
).drop(labels="PCode", axis=1)

df3
```
